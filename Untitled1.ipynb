```python
import sys
sys.path.append('..')

import numpy as np
import socnet as sn

from random import randint
```

```python
sn.node_size = 3
sn.node_color = (0, 0, 0)
sn.edge_width = 1
sn.edge_color = (192, 192, 192)
sn.node_label_position = 'top center'
    
gkiss = sn.load_graph('kiss-directed.gml')

sn.show_graph(gkiss, nlab=True)
```

```python
k = 10

scale = 0.8

residue = (1 - scale) / gkiss.number_of_nodes()

# inicializa ranks

rank = 1 / gkiss.number_of_nodes()

for n in gkiss.nodes():
    gkiss.node[n]['r'] = rank

for _ in range(k):
    # calcula quanto cada nó recebe

    received = {n: 0 for n in gkiss.nodes()}

    for n in gkiss.nodes():
        sent = gkiss.node[n]['r'] / len(gkiss.successors(n))

        for m in gkiss.successors(n):
            received[m] += sent

    # atualiza quanto cada nó tem, considerando escala e resíduo

    for n in gkiss.nodes():
        gkiss.node[n]['r'] = scale * received[n] + residue

# imprime ranks

for n in gkiss.nodes():
    print('{}: rank {:04.2f}'.format(gkiss.node[n]['label'], gkiss.node[n]['r'] * gkiss.number_of_nodes()))
```

```python
k = 10

# constrói matriz de adjacência
A = sn.build_matrix(gd_ts)

# constrói matriz transposta
At = A.transpose()

# inicializa arbitrariamente hubs
h = np.array([[0]] * gd_ts.number_of_nodes())

# inicializa arbitrariamente authorities
_a = []
for _ in range(gd_ts.number_of_nodes()):
    _a.append([randint(0, 10)])
a = np.array(_a)

for _ in range(k):
    # atualiza hubs a partir de authorities
    h = A.dot(a)

    # atualiza authorities a partir de hubs
    a = At.dot(h)

# soma hubs
sh = np.sum(h)

# soma authorities
sa = np.sum(a)

# imprime hubs e authorities normalizados
for n in gd_ts.nodes():
    label = gd_ts.node[n]['label']
    print('{}: hub {:04.3f}, authority {:04.3f}'.format(label, h[n, 0] / sh, a[n, 0] / sa))
```
